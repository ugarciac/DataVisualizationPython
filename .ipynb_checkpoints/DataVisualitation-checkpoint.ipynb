```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# paths
dataset='iris'

dataset_original = 'datasets'+'/'+dataset+'/'+dataset+'_original.csv'
dataset_with_noise = 'datasets'+'/'+dataset+'/'+dataset+'_with_noise.csv'
dataset_without_noise = 'datasets'+'/'+dataset+'/'+dataset+'_not_noise.csv'
```

```python
# original distribution
dataset_original = pd.read_csv(dataset_original)

# resulting distribution with noise
dataset_with_noise = pd.read_csv(dataset_with_noise)

# resulting distribution without noise
dataset_without_noise = pd.read_csv(dataset_without_noise)

# list of dataframes
datasets = [dataset_original, dataset_with_noise, dataset_without_noise]
```

```python
# Parameters
# colors
colors = ["tab:blue", "tab:green", "darkorange", "red"]

# axis (x,y)
x_var = ['sepal_length']
y_var = ['sepal_width']

# aditional parameters
hue='Class'
height=2.5
aspect=2.5
palette=colors
```

```python
# titles for figures
titles=['original '+dataset+' distribution','resulting distribution of '+dataset+' (with noise)','resulting distribution of '+dataset+' (without noise)']

for dataset,title in zip(datasets, titles):
    sns.pairplot(x_vars=x_var, y_vars=y_var, hue=hue, data=dataset, height=height, 
                 aspect=aspect, palette=palette).fig.suptitle(title)
```

```python
fig=sns.lmplot(x="sepal_length", y="sepal_width", hue=hue, data=dataset_with_noise, palette=palette)
```
