```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# paths
iris_original = 'datasets/iris/iris_original.csv'
iris_with_noise = 'datasets/iris/iris_with_noise.csv'
iris_not_noise = 'datasets/iris/iris_not_noise.csv'
```

```python
# original distribution
dataset_original = pd.read_csv(iris_original)

# resulting distribution with noise
dataset_with_noise = pd.read_csv(iris_with_noise)

# resulting distribution without noise
dataset_not_noise = pd.read_csv(iris_not_noise)

# list of dataframes
datasets = [dataset_original, dataset_with_noise, dataset_not_noise]
```

```python
# Parameters
# colors
colors = ["tab:blue", "tab:green", "darkorange", "red"]

# axis (x,y)
x_var = ['sepal_length']
y_var = ['sepal_width']

# aditional parameters
hue='Class'
height=2.5
aspect=2.5
palette=colors
```

```python
titles=['original iris distribution','resulting distribution of iris (with noise)','resulting distribution of iris (without noise)']
for dataset,title in zip(datasets, tittles):
    sns.pairplot(x_vars=x_var, y_vars=y_var, hue=hue, data=dataset, height=height, 
                 aspect=aspect, palette=palette).fig.suptitle(tittle)
```

```python
fig=sns.lmplot(x="sepal_length", y="sepal_width", hue=hue, data=dataset_with_noise, palette=palette)
```
